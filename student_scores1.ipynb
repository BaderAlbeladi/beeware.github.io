**Unlocking Student Potential: A Data Scientist's Journey Through Academic Insights**

---

As a data scientist, I'm often entrusted with raw datasets that hold untapped stories waiting to be unearthed. Recently, I delved into a fascinating dataset encompassing various aspects of student life and academic performance. This dataset included attributes like gender, part-time job status, absence days, extracurricular activities, weekly self-study hours, career aspirations, and scores across several subjects such as math, history, physics, chemistry, biology, English, and geography.

**The Quest Begins: Understanding the Dataset**

The first step was to familiarize myself with the data's landscape. Stripping away personal identifiers to maintain confidentiality, I focused on the variables that could shed light on students' academic journeys. The dataset wasn't just numbers and categories; it was a mosaic of student experiences, efforts, and aspirations.

**Painting the Big Picture: Descriptive Analysis**

I started with a descriptive analysis to capture the essence of the data. The gender distribution revealed a nearly balanced cohort, providing a fair ground for comparative insights. The majority of students engaged in extracurricular activities, and a significant portion held part-time jobs. Absence days varied widely, hinting at differing levels of engagement or external commitments.

When examining academic performance, I noticed that average scores differed across subjects. Math and physics had lower average scores compared to English and biology. Could this indicate that students found STEM subjects more challenging, or was there a deeper story?

**Connecting the Dots: Correlation and Patterns**

To uncover relationships between variables, I conducted a correlation analysis. Several compelling patterns emerged:

- **Self-Study Hours Matter:** There was a strong positive correlation between weekly self-study hours and scores in math and physics. Students dedicating more time to independent study tended to excel in these subjects.

- **The Impact of Absences:** Absence days negatively correlated with scores across all subjects. Frequent absences seemed to hinder academic performance, underscoring the importance of consistent attendance.

- **Extracurricular Activities:** Interestingly, participation in extracurricular activities showed a slight positive correlation with English and history scores. Perhaps these activities enhanced skills relevant to these subjects, like communication and critical thinking.

- **Part-Time Jobs:** Students with part-time jobs had mixed outcomes. While some managed their time effectively, others showed lower scores, possibly due to divided attention between work and studies.

**Diving Deeper: Subject-Specific Insights**

*Math and Physics:*

- **Challenge Accepted:** Students generally scored lower in math and physics. The complexity of these subjects likely required more dedicated study time.

- **Study Habits:** Those excelling in math and physics reported higher self-study hours, emphasizing the need for practice in these disciplines.

*English and History:*

- **Natural Affinities:** Higher average scores suggested that students were more comfortable with these subjects.

- **Extracurricular Influence:** Engagement in activities like debate clubs or writing workshops may have bolstered their skills in humanities.

**Unveiling Aspirations: The Career Connection**

Exploring career aspirations revealed insightful trends:

- **STEM Aspirants:** Students aiming for STEM careers showed higher engagement in self-study and performed better in related subjects, despite the inherent difficulties.

- **Arts and Humanities Pursuits:** Those interested in arts or social sciences excelled in English and history, aligning their academic strengths with future goals.

**The Human Element: Gender Dynamics**

Analyzing performance through the lens of gender brought additional nuances:

- **Closing the Gap:** While historical data often shows gender disparities in STEM fields, this dataset revealed a narrowing gap. Female students performed on par with their male counterparts in math and science.

- **Empowering Education:** The data suggested that ongoing efforts to encourage diversity in STEM might be bearing fruit, highlighting the importance of supportive educational environments.

**Challenges Uncovered: The Role of Absenteeism**

A significant finding was the detrimental effect of absenteeism on academic performance:

- **Missed Opportunities:** Each absence represented missed instruction, leading to knowledge gaps.

- **Underlying Causes:** Possible reasons for absences could range from health issues to disengagement or external responsibilities like part-time jobs.

2. Data Cleaning and Preparation
a. Handling Personal Identifiable Information (PII)
To protect student privacy, we'll exclude columns containing personal identifiable information such as id, first_name, last_name, and email from our analysis.**bold text**

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('student-scores.csv')  # Replace with your actual file name

# Preview the first few rows
print("First few rows of the dataset:")
display(df.head())

# Check the shape of the data
print(f"\nDataset contains {df.shape[0]} rows and {df.shape[1]} columns.")

# Print the column names to verify they exist
print("\nColumns in the DataFrame:")
print(df.columns)


# Drop PII columns
df.drop(['id', 'first_name', 'last_name', 'email'], axis=1, inplace=True)
```

```python
# Drop PII columns
#df.drop(['id', 'first_name', 'last_name', 'email'], axis=1, inplace=True)
```

# b. Checking for Missing Values
# We'll check for any missing values and decide how to handle them.

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

# Handle missing values if necessary
# For simplicity, let's drop rows with missing values
df.dropna(inplace=True)
```

# c. Data Type Conversion
Ensure that all numerical columns are correctly formatted.

```python
# List of score columns
score_columns = ['math_score', 'history_score', 'physics_score', 'chemistry_score',
                 'biology_score', 'english_score', 'geography_score']

# Convert score columns to numeric
for col in score_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert categorical variables to appropriate types
categorical_cols = ['gender', 'part_time_job', 'extracurricular_activities', 'career_aspiration']
df[categorical_cols] = df[categorical_cols].astype('category')
```

## d. Rechecking Data

```python
# Verify data types
print("\nData types after conversion:")
print(df.dtypes)

# Verify dataset shape after cleaning
print(f"\nDataset now contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

# 3. Exploratory Data Analysis (EDA)
# a. Descriptive Statistics
# We'll compute basic statistics to understand the distribution of our data.

```python
# Descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical variables:")
display(df.describe())

# Descriptive statistics for categorical columns
print("\nDistribution of categorical variables:")
for col in categorical_cols:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())
```

# b. Distribution of Scores
# Assessing the distribution of scores in each subject.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# Histograms for each subject score
for col in score_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=10, color='skyblue')
    plt.title(f'Distribution of {col.replace("_", " ").capitalize()}')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.show()
```

# c. Gender Distribution


```python
# Count plot for gender
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# 4. Statistical Analysis
# a. Correlation Analysis
# Understanding relationships between variables.

```python
# Encode categorical variables for correlation analysis
df_encoded = df.copy()
for col in categorical_cols:
    df_encoded[col] = df_encoded[col].cat.codes

# Compute the correlation matrix
corr_matrix = df_encoded.corr()

# Display the correlation matrix
print("\nCorrelation matrix:")
display(corr_matrix)
```

## b. Heatmap of Correlations

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
